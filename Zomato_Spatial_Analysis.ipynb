```python
df.dropna(inplace=True,subset=['location'])
```

```python
df['location'] = df['location'] +', Bangalore, Karnataka, India'
```

```python
df['location'].head(10)
```

```python
location = pd.DataFrame()
location['Name'] = pd.DataFrame(df['location'].unique())
```

```python
location
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="app",timeout=None)

# Add new columns for latitude and longitude
location['Lat'] = None
location['Long'] = None

for i, row in location.iterrows():
    loc = geolocator.geocode(row['Name'])
    if loc is not None:
        location.at[i, 'Lat'] = loc.latitude
        location.at[i, 'Long'] = loc.longitude
```

```python
location.at[79,'Lat'] = 13.0120218
location.at[79,'Long'] = 77.6777817
location.at[85,'Lat'] = 13.010316
location.at[85,'Long'] = 77.580569
```

```python
Rest_loc= df['location'].value_counts().reset_index()
Rest_loc.columns = ['Name','Count']
```

```python
new_df = Rest_loc.merge(location,on='Name')
```

```python
new_df.head(10)
```

# HeatMap


```python
import folium
def generate_basemap():
  basemap=folium.Map(location=[12.971599,77.594566])
  return basemap
```

```python
basemap= generate_basemap()
```

```python
from folium.plugins import HeatMap
HeatMap(data=new_df[['Lat','Long','Count']]).add_to(basemap)
# Add a title to the map
title_html = '''
     <h3 align="center" style="font-size:20px"><b>Restaurant Count HeatMap</b></h3>
     '''
basemap.get_root().html.add_child(folium.Element(title_html))
basemap
```

# Marker Cluster


```python
from folium.plugins import FastMarkerCluster
# Convert DataFrame columns to lists
basemap2= generate_basemap()

FastMarkerCluster(data=new_df[['Lat','Long','Count']]).add_to(basemap2)
# Add a title to the map
title_html = '''
     <h3 align="center" style="font-size:20px"><b>Restaurant Count Marker Cluster</b></h3>
     '''
basemap2.get_root().html.add_child(folium.Element(title_html))
basemap2
```

```python
basemap3= generate_basemap()
for i,row in new_df.iterrows():
  folium.Marker(location=[row['Lat'],row['Long']],popup=row['Name']).add_to(basemap3)

# Add a title to the map
title_html = '''
     <h3 align="center" style="font-size:20px"><b>Restaurant Coordinate with Popup Location Name</b></h3>
     '''
basemap3.get_root().html.add_child(folium.Element(title_html))
basemap3
```

# Data Cleaning and Transformation


```python
df['rate'].isnull().sum()
```

```python
df.dropna(inplace=True,subset=['rate'])
```

```python
df['rate'].unique()
```

```python
df['rate'].replace('NEW','0',inplace=True)
df['rate'].replace('-','0',inplace=True)
```

```python
df['rating']=df['rate'].str.replace('/5','')
```

```python
df['rating'].unique()
```

```python
df['rating']=df['rating'].astype(float)
```

```python
group_df=df.groupby('location',as_index=False).agg({'rating':'mean','name':'count'})
```

```python
group_df.columns=['Name','Avg Rating','Count']
```

```python
group_df.head()
```

```python
group_df= group_df.merge(new_df.drop('Count',axis=1),on='Name')
```

# Average Rating Plot


```python
basemap4= generate_basemap()
HeatMap(data=group_df[['Lat','Long','Avg Rating']]).add_to(basemap4)
# Add a title to the map
title_html = '''
     <h3 align="center" style="font-size:20px"><b>Restaurant Average Rating By Coordinates</b></h3>
     '''
basemap4.get_root().html.add_child(folium.Element(title_html))
basemap4
```
